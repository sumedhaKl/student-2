## SASS: Syntactically Awesome Stylesheets

![SASS Logo](https://upload.wikimedia.org/wikipedia/commons/thumb/9/96/Sass_Logo_Color.svg/1024px-Sass_Logo_Color.svg.png)

### What is SASS?

SASS is a preprocessor scripting language that is interpreted or compiled into CSS. It takes input data and converts it into CSS, which is then used by the browser.

### Understanding differences: CSS V/S SASS

- **CSS:** Cascading style sheets or CSS is the standard styling language for the web, providing a simple and direct syntax for styling HTML elements. However, it lacks certain features found in preprocessors like Sass, making it less understandable and difficult to work with.

```css
/* Basic CSS Code */

body {
  font-family: Arial, sans-serif;
  background-color: #f0f0f0;
}

header {
  background-color: #333;
  color: white;
  padding: 10px;
}
```

- **SASS:** SASS is a powerful preprocessor scripting language that enhances CSS with advanced features. Sass introduces variables, nesting, and mixins, requiring compilation into standard CSS for browsers to interpret.

```scss
/* Basic SASS Code */

$font-family: Arial, sans-serif;
$background-color: #f0f0f0;
$main-color: #333;

body {
  font-family: $font-family;
  background-color: $background-color;
}

header {
  background-color: $main-color;
  color: white;
  padding: 10px;
}
```

## Benefits of SASS over CSS

1. **Improved Code Organization:**
   - Nesting Capabilities: Sass organizes styles like a neat notebook, mirroring HTML structure for clarity.
   - Modular Approach: Break styles into smaller files, like building blocks, for a cleaner and more organized project.

2. **Maintainability:**
   - Use of Variables: Easily update colors and fonts with name-tag-like variables.
   - Mixins for Reusable Code: Create styles once, reuse them everywhere.
   - Extend/Inheritance: Share styles among elements to avoid repetition.

3. **Advanced Features:**
   - Control Directives: Add conditions to styles for more flexibility.
   - Built-in Functions: Use Sass tools for colors and math in styles.
   - Compatibility: Sass handles browser differences, ensuring a consistent look across all browsers.

4. **Preprocessing:**
   - Compilation: Translate Sass to standard CSS for browsers.
   - Usage: Command Sass to compile styles effortlessly.
   - Watching Files: Sass auto-updates styles as you make changes, simplifying the process.

```scss
// Variables
$firstName: "John";
$lastName: "Doe";

// String concatenation using whitespace
$fullName: $firstName + " " + $lastName;

#testing {
  content: $fullName;
}
```

**Intended Answer:**
```
content: John Doe
```

2. **Conditional Statements:**
   - Conditional statements in SASS work and function the same way they do in Python and Javascript.

```scss
// Example code

$grade: 85;

.button {
  @if $grade >= 90 {
    background-color: green;
    color: white;
    border: 2px solid darkgreen;
  } @else if $grade >= 80 {
    background-color: yellow;
    color: darkgray;
    border: 2px solid goldenrod;
  } @else {
    background-color: red;
    color: white;
    border: 2px solid darkred;
  }
}
```

3. **Loops in SASS:**
   - In SASS, loops are facilitated by the `@for`, `@while`, and `@each` decorators. Loops enable the repetition of a code block either for a specified number of iterations or until a particular condition is satisfied.

```scss
// Example code

$alignments: left, center, right;

@each $alignment in $alignments {
  .text-#{$alignment} {
    text-align: $alignment;
  }
}
```

**Expected output:**

```css
.text-left {
  text-align: left;
}

.text-center {
  text-align: center;
}

.text-right {
  text-align: right;
}
```

### <h2>How does HTML code use SASS??</h2>

In HTML, you don't directly use Sass; instead, you use the compiled CSS generated from Sass.

#### <h4>Steps to use predefined SASS</h4>

1. Install SASS: Before you use SASS you need to install it.

```bash
npm install -g sass
```

2. Create SASS file: Write your SASS file code and be sure to give it the extension of ".scss"

```scss
// Example sass

$primary-color: #3498db;

body {
  background-color: $primary-color;
}
```

3. Compile SASS into CSS: Use installed SASS compiler to convert the .scss file into a .css file.

```bash
sass input.scss output.css
```

4. Link compiled CSS into the HTML/JS project code

```html
<html>
<head>
  <!--Meta data here -->
  <link rel="stylesheet" href="output.css">
  <title>Project Page</title>
</head>
<body>
  <!--HTML stuff-->
</body>
</html>
```
